Confusion Matrix for multiclass classification

```python
import matplotlib.pyplot as plt
import numpy
from sklearn import metrics

actual = results["actual"].values.tolist()  # actual values for based on both side
predicted = results["pred"].values.tolist()      # predicted values for based on both side

confusion_matrix = metrics.confusion_matrix(actual, predicted)

cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = ['0', '2a', '2b', '3a', '3b', '4'])

cm_display.plot()
plt.show()
```

```python
actual_right_side_nonflip = results["actual_right_side"].values.tolist()  # actual values for right side
actual_right_side_flip = results["actual_flip_right_side"].values.tolist()  # actual values for flip right side
actual_sidebased = actual_right_side_nonflip + actual_right_side_flip

predicted_right_side_nonflip = results["pred_right_side"].values.tolist()      # predicted values for right side
predicted_right_side_flip = results["pred_flip_right_side"].values.tolist()      # predicted values for flip right side
predicted_sidebased = predicted_right_side_nonflip + predicted_right_side_flip


confusion_matrix = metrics.confusion_matrix(actual_sidebased, predicted_sidebased)

cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = ['0', '2', '3' ])

cm_display.plot()
plt.show()
```
