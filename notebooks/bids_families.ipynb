```python
from BIDS import BIDS_Global_info
import pandas as pd

WORKING_DIR = "/data1/practical-sose23/"

dataset = [WORKING_DIR + 'castellvi/3D-Castellvi-Prediction/data/dataset-verse19',  
               WORKING_DIR + 'castellvi/3D-Castellvi-Prediction/data/dataset-verse20', 
               WORKING_DIR + 'castellvi/3D-Castellvi-Prediction/data/dataset-tri']

image_types = ["ct", "subreg", "cortex"]

data_types = ['rawdata',"derivatives"]

bids = BIDS_Global_info(dataset, data_types, additional_key = image_types, verbose=True)
```

```python
len(bids.subjects)
data_dict =  {}
master_list = "/data1/practical-sose23/castellvi/3D-Castellvi-Prediction/data/Castellvi_list.xlsx"
master_df = pd.read_excel(master_list)
subject_list = master_df.dropna(subset = "Full_Id")['Full_Id'].values.tolist()

for subject in subject_list:
    decomposed_id = subject.split("_")
    data_dict[decomposed_id[0]] = {}
    data_dict[decomposed_id[0]]["split"] = None if len(decomposed_id) == 1 else decomposed_id[1]

for subject in data_dict.keys():

         
    subject_container = bids.subjects[subject[4:]]
    query = subject_container.new_query(flatten = True)
    if subject[4:9] == "verse":
        dataset = "verse"
    elif subject[4:7] == "tri":
        dataset = "tri"

    split = data_dict[subject]["split"] 

    try:
        raw_file = None
        seg_file = None
        ctd_file = None

        for file in query.loop_list():
            
            split = True if data_dict[subject]["split"] == None else file.do_filter("split", data_dict[subject]["split"][6:])

            if file.do_filter("format", "ct") and split:
                raw_file = file
            elif (file.do_filter("seg", "subreg") or file.do_filter("seg", "subregsac")) and file.do_filter("format", "msk") and split:
                seg_file = file
            if file.do_filter("format", "ctd") and (file.do_filter("seg", "subreg") or file.do_filter("seg", "subregsac")) and split:
                ctd_file = file

        assert(raw_file != None)
        assert(seg_file != None)
        assert(ctd_file != None)

        data_dict[subject]["raw_file"] = raw_file
        data_dict[subject]["seg_file"] = seg_file
        data_dict[subject]["ctd_file"] = ctd_file
            
    except Exception as e:
        print("Subject {} threw Error {}".format(subject, e))
        
        
data_dict["sub-verse400"]
```

```python
def split_full_id(full_id: str):
    decomposed_id = full_id.split("_")
    sub = decomposed_id[0]
    if sub[:4] == "sub-":
        #clean Full_Id prefixes
        sub = sub[4:]
    if sub[:5] == "verse":
        dataset = "verse"
        split = None if len(decomposed_id) == 1 else decomposed_id[1][6:]
        ce = None
    elif sub[:3] == "tri":
        dataset = "tri"
        split = None
        ce = None if len(decomposed_id) == 1 else decomposed_id[1][3:]
    else:
        raise Exception("Unrecognised dataset {}".format(full_id))
    
    return sub, dataset, split, ce

def create_verse_record(sub, split):
    record = {}

    record["dataset"] = "verse"

    record["subject"] = sub
    record["split"] = split

    raw_file = None
    seg_file = None
    ctd_file = None

    subject_container = bids.subjects[sub]
    query = subject_container.new_query(flatten = True)
    
    for file in query.loop_list():

        split_filter = True if split == None else file.do_filter("split", split)

        if file.get_parent() == "rawdata" and split_filter:
            raw_file = file
        elif (file.do_filter("seg", "subreg") or file.do_filter("seg", "subregsac")) and file.do_filter("format", "msk") and split_filter:
            seg_file = file
        if file.do_filter("format", "ctd") and (file.do_filter("seg", "subreg") or file.do_filter("seg", "subregsac")) and split_filter:
            ctd_file = file

    assert(raw_file != None)
    assert(seg_file != None)
    assert(ctd_file != None)

    record["raw_file"] = raw_file
    record["seg_file"] = seg_file
    record["ctd_file"] = ctd_file

    return record

def create_tri_record(sub, ce):
    record = {}
    record["dataset"] = "tri"

    record["subject"] = sub
    record["ce"] = ce

    raw_file = None
    seg_file = None
    ctd_file = None

    subject_container = bids.subjects[sub]
    query = subject_container.new_query(flatten = True)

    for file in query.loop_list():

        if file.get_parent() == "rawdata" and file.do_filter("ce", ce):
            raw_file = file
        elif (file.do_filter("seg", "subreg") or file.do_filter("seg", "subregsac")) and file.do_filter("format", "msk") and file.do_filter("ce", ce):
            seg_file = file
        if file.do_filter("format", "ctd") and (file.do_filter("seg", "subreg") or file.do_filter("seg", "subregsac")) and file.do_filter("ce", ce):
            ctd_file = file

    assert(raw_file != None)
    assert(seg_file != None)
    assert(ctd_file != None)

    record["raw_file"] = raw_file
    record["seg_file"] = seg_file
    record["ctd_file"] = ctd_file

    return record

    
```

```python
verse_records = []
tri_records = []

for record in master_records:
    sub, dataset, split, ce = split_full_id(record["Full_Id"])
    if dataset == "verse":
       verse_record = create_verse_record(sub, split)
       verse_records.append(verse_record)
    elif dataset == "tri":
        tri_record = create_tri_record(sub, ce)
        tri_records.append(tri_record)
       
```
